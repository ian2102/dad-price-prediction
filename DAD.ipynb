```python
import pandas as pd
import joblib
import numpy as np
from datetime import datetime
from sklearn.ensemble import RandomForestRegressor
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.pipeline import Pipeline
import scheme

data = pd.read_csv('data/2025-03-16.csv', encoding='utf-8', encoding_errors='ignore')
data = data.dropna()
```

```python
df = data
df = df.drop(columns=['text'])

df = pd.get_dummies(df)
item = scheme.get_empty_item()
missing_cols = list(set(item) - set(data.columns))
missing_df = pd.DataFrame(0, index=data.index, columns=missing_cols)
data = pd.concat([data, missing_df], axis=1)
    
X = df.drop(columns=['price'])
y = df['price']

model = RandomForestRegressor()
model.fit(X, y)

num_samples = int(len(df) / 1000)
model_filename = f"price_{num_samples}K_{datetime.now().date()}_RandomForestRegressor.joblib"
joblib.dump(model, "models/" + model_filename)

print(f"Model saved as {model_filename}")
```

```python
df = data
X = df["text"]
df = df.drop(columns=['text'])

df = pd.get_dummies(df)
item = scheme.get_empty_item()
missing_cols = list(set(item) - set(data.columns))
missing_df = pd.DataFrame(0, index=data.index, columns=missing_cols)
data = pd.concat([data, missing_df], axis=1)
    
y = df.drop(columns=['price'])

vectorizer = CountVectorizer()
rf_model = RandomForestRegressor()

pipeline = Pipeline([
    ('vectorizer', vectorizer),
    ('model', rf_model)
])

pipeline.fit(X, y)
num_samples = int(len(df) / 1000)
model_filename = f"image_{num_samples}K_{datetime.now().date()}_RandomForestRegressor.joblib"
joblib.dump(pipeline, "image_recognition/" + model_filename)

print(f"Model saved as {model_filename}")
```
